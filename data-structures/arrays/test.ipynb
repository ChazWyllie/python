```python
s = "IceCreAm"
s = list(s)
p1, p2 = 0, len(s) - 1
vowel = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
temp = []

for i in range(len(s)):
    if s[i] in vowel:
        temp.append(s[i])

while p1 < p2:
    if s[p1] not in vowel:
        p1 += 1
    elif s[p2] not in vowel:
        p2 -= 1
    else:
        s[p1], s[p2] = s[p2], s[p1]
        p1 += 1
        p2 -= 1
print(''.join(s))
```
